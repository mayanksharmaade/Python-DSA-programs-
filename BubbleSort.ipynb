## Bubble Sort

```python
## implementation of unsorted 1 D array using bubble sort 
```

```python
def Bubblesort(arr):
    n=len(arr)
    for i in range(n):
        for j in range (0,n-i-1):
            if arr[j]> arr[j+1]:
                ## swap the elements
                arr[j],arr[j+1]=arr[j+1],arr[j]
    return arr           
## Driver code
arr=[70,20,50,30,90,5,15]
result=Bubblesort(arr)
print('The sorted array is',result)
```
