## INSERTION SORT

```python
#Insertion Sort Compares the element at 1st Index and the 0th Index and swaps the values if the Key at 1st index is smaller 
#When a new element is compared this entire process is repeated and The new Element at next index is compared with the sorted array on the left 


```

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        ## holding the key to be compared with the left element 
         key = arr[i]
         j = i - 1
# Move elements of arr[0..i-1] that are greater than key
         while j >= 0 and key < arr[j]:
               arr[j + 1] = arr[j]
               j -= 1
         arr[j + 1] = key

# Example usage
data = [12, 11, 13, 5, 6]
insertion_sort(data)
print("Sorted array:", data)
```
