## SELECTION SORT 

```python
## Initialize the First element as the minimum 
## Compare with the next elements of the array , and swap  with the minimum element 
## Repeast the process untill entire array is sorted 

## Time Complexity : O(n^2) for best , worst, and average case
## Space Complexity : O(1)
```

```python
def selectionSort(array):
    
    size = len(array)
    for step in range(size):
        ## Selecting the minimum element 
       min_idx = step
       for i in range(step + 1, size):
          if array[i] < array[min_idx]:
               min_idx = i
              ## swapping the elements 
               array[step], array[min_idx] = array[min_idx], array[step]

# Example usage
data = [64, 25, 12, 22, 11,7,4,43]
selectionSort(data)
print('Sorted Array in Ascending Order:')
print(data)
```
