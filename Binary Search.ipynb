## Binary Search 

```python
## Binary search  finds the element in an sorted array and divides it into two parts  .  finds the middle element in an array and compares it with the Key .
## if the key is smaller , it looks for the elements in left part otherwise in the right part 

## Time Complexity : O(log n)
## Iterative Binary Serach 
```

```python
def BinarySearch(A,key) :
    l=0
    r=len(A)-1
   
    while l<=r:
        mid=(l+r) // 2
        
        if (key==A[mid]):
            return mid
        elif (key >A[mid]):
            l=mid+1
        else :
            r= mid-1
    return -1

## Calling Function

A=[12,17,19, 45,67, 75, 82]
result=BinarySearch(A,75)

print('Result:', result)

            
        
```

## Recursive Binary Search

```python
def BinarySearchRecursive( arr, i,j,x):
    while i<=j:
        mid=i+(j-i) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid]<x:
            return BinarySearchRecursive(arr,mid+1,j,x)
        else:
            return BinarySearchRecursive(arr,i,mid-1,x)
    return -1


 ## Driver Code 
## Sorted array
## Function Calling

arr=[2,5,10,14,18,22,27,35,40,45]
x=35
i=0
j=len(arr)-1

result=BinarySearchRecursive(arr,i,j,x)
print('the index is :',result)

    
```
