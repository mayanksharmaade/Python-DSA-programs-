## Linear search in Python

```python
# Finding an element from  the list , returns Item not found when element is not present 
numbers=[12,4,6,8,23,15,9]
find=int(input('input the number to be found'))
found= False
counter=1

for number in numbers:
    if number==find:
        found=True
        print('found at position',counter)
    else:
        counter=counter+1

if not found:
    print('item not found')
    
        




    
```
